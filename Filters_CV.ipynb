```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img = cv2.imread('/content/rose.jpg')
```

```
plt.figure(figsize=(20, 8))
plt.imshow(img)
```

```
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
plt.figure(figsize=(20, 8))
plt.imshow(image_RGB)
```

Average Filter

```
blurred_image = cv2.blur(image_RGB, (50, 50))
```

```
plt.figure(figsize = (20, 8))
plt.imshow(blurred_image)
```

Median Filter

```
median_blur = cv2.medianBlur(image_RGB,51)
```

```
plt.figure(figsize = (20, 8))
plt.imshow(median_blur)
```

Gaussian Filter

```
gaussian_blur = cv2.GaussianBlur(image_RGB, (51, 51), 0)
```

```
plt.figure(figsize = (20, 8))
plt.imshow(gaussian_blur)
```

Bilateral Filter

```
bilateral_blur = cv2.bilateralFilter(image_RGB, 23, 51, 51)
```

```
plt.figure(figsize = (20, 8))
plt.imshow(bilateral_blur)
```

Difference in noise in the original image and output image of bilateral filter

```
original_noise = cv2.Canny(image_RGB, 2, 15)
plt.figure(figsize = (20, 8))
plt.imshow(original_noise)
```

```
bilateral_noise = cv2.Canny(bilateral_blur, 2, 15)
plt.figure(figsize = (20, 8))
plt.imshow(bilateral_noise)
```
